```
from snowflake.sqlalchemy import URL
from sqlalchemy import create_engine

user = dbutils.secrets.get("data-warehouse", "<snowflake-user>")
password = dbutils.secrets.get("data-warehouse", "<snowflake-password>")

engine = create_engine(URL(
    user=user,
    password=password,
    account='sqa35008.us-east-1',
    warehouse='INTERVIEW_WH',
    database='USER_NW',
    schema='PUBLIC',
    role='SYSADMIN',
))

connection = engine.connect()
```

```
try:
  connection.execute(dbutils.widgets.get("AIRLINES"))
  connection.execute(dbutils.widgets.get("AIRPORTS"))
  connection.execute(dbutils.widgets.get("FLIGHTS"))
finally:
  connection.close()
  engine.dispose()
```
